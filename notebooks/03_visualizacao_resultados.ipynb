```python
# ===========================================================================
# CÉLULA 2 — Importações e Configurações de Estilo
# ===========================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from pathlib import Path

# Estilo para publicação científica
plt.rcParams.update({
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'font.family': 'DejaVu Sans',
    'font.size': 11,
    'axes.titlesize': 13,
    'axes.labelsize': 11,
    'axes.spines.top': False,
    'axes.spines.right': False,
})
sns.set_palette('colorblind')

FIGURES_DIR = Path('../results/figures')
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

print(f'✅ Figuras serão salvas em: {FIGURES_DIR}')
```

```python
# ===========================================================================
# CÉLULA 3 — Carregamento dos Dados Analisados
# ===========================================================================
INPUT_PATH = Path('../data/processed/papers_analisados.csv')

if INPUT_PATH.exists():
    df = pd.read_csv(INPUT_PATH, encoding='utf-8')
    print(f'✅ Dados carregados: {len(df)} artigos')
else:
    print('⚠️  Execute primeiro os notebooks 01 e 02.')
    # Dados de exemplo
    np.random.seed(42)
    years = np.random.choice(range(2020, 2027), 50)
    venues = np.random.choice(['Computers & Education', 'AI & Society', 'AIED', 'Nature', 'PLOS ONE'], 50)
    df = pd.DataFrame({
        'paperId': [f'id_{i}' for i in range(50)],
        'year': years,
        'venue': venues,
        'citationCount': np.random.randint(0, 200, 50),
        'cluster': np.random.randint(0, 5, 50),
        'flag_impacto': np.random.choice([True, False], 50),
        'flag_desigualdade': np.random.choice([True, False], 50),
        'flag_etica': np.random.choice([True, False], 50),
        'flag_ia_generativa': np.random.choice([True, False], 50),
    })
    print(f'📋 Usando dados de exemplo: {len(df)} artigos')
```

```python
# ===========================================================================
# CÉLULA 4 — Figura 1: Distribuição Temporal de Publicações
# ===========================================================================
fig, ax = plt.subplots(figsize=(10, 5))

year_counts = df['year'].value_counts().sort_index()

ax.bar(
    year_counts.index,
    year_counts.values,
    color=sns.color_palette('colorblind')[0],
    edgecolor='white',
    linewidth=0.8,
    alpha=0.9
)

ax.set_title('Distribuição Temporal de Publicações sobre IA na Educação (2020–2026)', 
             fontweight='bold', pad=15)
ax.set_xlabel('Ano de Publicação')
ax.set_ylabel('Número de Artigos')
ax.xaxis.set_major_locator(mticker.MultipleLocator(1))

for bar in ax.patches:
    ax.annotate(
        str(int(bar.get_height())),
        (bar.get_x() + bar.get_width() / 2, bar.get_height()),
        ha='center', va='bottom', fontsize=9, fontweight='bold'
    )

plt.tight_layout()
fig.savefig(FIGURES_DIR / 'fig1_distribuicao_temporal.png', bbox_inches='tight')
plt.show()
print('✅ Figura 1 salva.')
```

```python
# ===========================================================================
# CÉLULA 5 — Figura 2: Top 10 Periódicos por Número de Publicações
# ===========================================================================
fig, ax = plt.subplots(figsize=(10, 6))

top_venues = df['venue'].value_counts().nlargest(10)

ax.barh(
    top_venues.index[::-1],
    top_venues.values[::-1],
    color=sns.color_palette('colorblind')[1],
    edgecolor='white',
    alpha=0.9
)

ax.set_title('Top 10 Periódicos — Volume de Publicações sobre IA na Educação',
             fontweight='bold', pad=15)
ax.set_xlabel('Número de Artigos')
ax.set_ylabel('')

plt.tight_layout()
fig.savefig(FIGURES_DIR / 'fig2_top_periodicos.png', bbox_inches='tight')
plt.show()
print('✅ Figura 2 salva.')
```

```python
# ===========================================================================
# CÉLULA 6 — Figura 3: Mapa de Calor — Clusters × Flags Sociológicas
# ===========================================================================
flag_cols = [c for c in df.columns if c.startswith('flag_')]

if flag_cols and 'cluster' in df.columns:
    heatmap_data = df.groupby('cluster')[flag_cols].mean().T
    heatmap_data.index = [c.replace('flag_', '').replace('_', ' ').title() for c in heatmap_data.index]

    fig, ax = plt.subplots(figsize=(10, 5))
    sns.heatmap(
        heatmap_data,
        annot=True,
        fmt='.2f',
        cmap='YlOrRd',
        linewidths=0.5,
        ax=ax,
        vmin=0, vmax=1
    )
    ax.set_title('Proporção de Artigos por Cluster e Categoria Temática',
                 fontweight='bold', pad=15)
    ax.set_xlabel('Cluster Temático')
    ax.set_ylabel('Categoria')

    plt.tight_layout()
    fig.savefig(FIGURES_DIR / 'fig3_heatmap_clusters.png', bbox_inches='tight')
    plt.show()
    print('✅ Figura 3 salva.')
else:
    print('⚠️  Colunas de flags ou cluster não encontradas.')
```

```python
# ===========================================================================
# CÉLULA 7 — Sumário Final
# ===========================================================================
print('=' * 60)
print('SUMÁRIO DAS VISUALIZAÇÕES GERADAS')
print('=' * 60)
for fig_file in sorted(FIGURES_DIR.glob('*.png')):
    size_kb = fig_file.stat().st_size / 1024
    print(f'  📊 {fig_file.name} ({size_kb:.1f} KB)')
print('=' * 60)
print(f'  Diretório: {FIGURES_DIR.resolve()}')
print(f'  Resolução: 300 DPI (pronto para publicação)')
```
